# Question3

Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where
s = (a+b+c)/2. 

```python

    class Program
    {
        static void Main(string[] args)
        {
            double a = 0, b = 0, c = 0;
            Console.WriteLine("Enter the value of first side of the triangle: ");
            a = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter the value of second side of the triangle: ");
            b = Convert.ToDouble(Console.ReadLine());
            Console.WriteLine("Enter the value of third side of the triangle: ");
            c = Convert.ToDouble(Console.ReadLine());
            #function call
            double area = GetArea(a, b, c);
            Console.WriteLine("The area of the triangle by heron's formula is:" + area);

        }
        #function definition
        static double GetArea(double a, double b, double c)
        {
            double s = 0, temp = 0;
            s = (a + b + c) / 2;
            temp = s * (s - a) * (s - b) * (s - c);
            return Math.Sqrt(temp);
        }
    }
```

• If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
• If the word starts with a vowel, simply append 'way' to the end of the word. For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case
characters. Your output should always be lower case however. 

```python
def pig(c):
    vowels = ['a','e','i','o','u']
    temp1="ay"
    temp2 = "way"
    if c[0] in vowels:
        print( c + temp2)
    else:
        print (c[1:] + c[0] + temp1)
s = input("Enter the String: ")
pig(s)
```
